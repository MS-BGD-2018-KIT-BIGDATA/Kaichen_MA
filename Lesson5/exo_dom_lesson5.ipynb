## Recap 5

#### Peut-on établir un lien entre la densité de médecins par spécialité  et par territoire et la pratique du dépassement d'honoraires ? 

#### Est-ce  dans les territoires où la densité est la plus forte que les médecins  pratiquent le moins les dépassement d'honoraires ? 

#### Est ce que la densité de certains médecins / praticiens est corrélé à la densité de population pour certaines classes d'ages (bebe/pediatre, personnes agées / infirmiers etc...) ?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_densite = pd.read_excel("Effectif_et_densite_par_departement_en_2014.xls",sheetname="Spécialistes",na_values ="nc")
df_densite = df_densite.iloc[:,0:5]
df_densite.head()
```

```python
df_honoraire = pd.read_excel("Honoraires_totaux_des_professionnels_de_sante_par_departement_en_2014.xls",sheetname ="Spécialistes",na_values ="nc")
df_honoraire.rename(columns={"EFFECTIFS": "EFFECTIF"},inplace=True)
df_honoraire.head()
```

```python
df = pd.merge(df_densite, df_honoraire,on=["Spécialistes","DEPARTEMENT","EFFECTIF"])
df.head()
```

```python
df_region = df.groupby("DEPARTEMENT").sum()
df_region["DEPASSEMENTS /Docteur (Euros)"]= df_region["DEPASSEMENTS (Euros)"]/df_region["EFFECTIF"]
df_region.sort_values("DENSITE /100 000 hab." , ascending=True).head()
```

```python
fig = plt.figure(figsize=(10,6))
plt.xlabel('DENSITE DE DOCTEURS"]')
plt.ylabel('DEPASSEMENTS / DOCTEUR (Euros)]')
plt.title('Lien entre dépassements d\'honoraires et la densité des medecins ')
plt.scatter(df_region["DENSITE /100 000 hab."], df_region["DEPASSEMENTS /Docteur (Euros)"])
plt.show()
```

#### Selon l'analyse au dessus, il est trouvé que plus la densité de medecins est élevée, plus le montant de dépassement est grave. Ce phénomène est possible vient de le faible occurence d'accuilir des patients pour les département sont plus medecins.
